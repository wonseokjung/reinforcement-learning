```python
#폴리시 이벨류에이션 함수 생성

def policy_eval(policy, env, discount_factor=1.0, theta=0.00001): #아규먼트는 폴리시, 환경, 디스카운트팩터, 세타값을 받는다
    """
    Evaluate a policy given an environment and a full description of the environment's dynamics.
    
    Args:
    폴리시 : 스테이트와 엑션으로 이루어진 메트릭스 
    엑션 : 오픈에이 아이의 환경,  env.P는 환경에서 트렌지션 확률을 의미한다.env.P[s][a] 은 (확률, 다음 스테이트, 리워드, 끝) 의 터플이다. 
    세타 : 관찰값의 변화가 이 세타값보다 작으면 멈춘다. 
    디스카운트펙터 -> 현재가 아닌 미래의 리워드의 기대값을 줄여주기  위한 하이퍼파라미터
    
    리턴: 
        벨류 함수를 나타내는 16개의 스테이트를 나타내는 벡터
    """
    
    #값이 0인 벨류함수로 시작
    

    V = np.zeros(env.nS) 
    
    
    
    while True:#While looping until True
        
        delta = 0 #델타값을 0으로 지정
        
        for s in range(env.nS): #모든 스테이트에서 ("full backup")
            
            v = 0 #벨류의 초기값은 0으로 설정
            # Look at the possible next actions
            for a, action_prob in enumerate(policy[s]): 
                
                
                #가능한 엑션과 그 엑션을 할 확률                
                for  prob, next_state, reward, done in env.P[s][a]: #env.P[s][a]에 들어있는 확률, 다음스테이트, 리워드, 끝을 포문으로 돌림
                    # 벨류값 계산
                    v += action_prob * prob * (reward + discount_factor * V[next_state])
            #value function이 얼마나 변했는지 계산
            delta = max(delta, np.abs(v - V[s]))
            V[s] = v 
            
            print(V[s])
            
        # Stop evaluating once our value function change is below a threshold
        #정해놓은 세타값보다 delta가 작으면 멈춘다. 
         
        if delta < theta:
            break
    return np.array(V)
```

```python


#0.25의 확률을 가지고 있는 엑션들을 생성
random_policy = np.ones([env.nS, env.nA]) / env.nA


v = policy_eval(random_policy, env)
```

```python
print("Value Function:")
print(v)
print("")

print("Reshaped Grid Value Function:")
print(v.reshape(env.shape))
print("")
```
